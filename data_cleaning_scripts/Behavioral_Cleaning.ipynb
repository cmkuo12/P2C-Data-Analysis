```python
from google.colab import drive
import pandas as pd
import json
from pandas.io.json import json_normalize
import datetime
import os

drive.mount('/content/drive', force_remount=True)
```

```python
def clean(path, json_file, output):
  json_dict = {}

  with open(path + json_file, 'r') as j:
      contents = json.loads(j.read())

  df = pd.json_normalize(contents)

  for column in df:
      date = datetime.datetime.fromtimestamp(int(column) / 1e3)
      arr = (df[column].to_numpy())[0]
      for key in arr:
        if key[0] == "poseDetected" or key[0] == "codestate" or key[0] == "completedExercise" or key[0] == "runCode":
          value_map = {key[0] : key[1]}
          json_dict[str(date)] = value_map

  print(json_dict)

  f = open(output + "CleanJSON/" + json_file, "w")
  json.dump(json_dict, f)
  f.close()
```

```python
path = "/content/drive/MyDrive/PoseToCode - Top Level Folder/RealWorldStudyData/firebase_data/"
for filename in os.listdir(path):
    clean(path, filename, "/content/drive/MyDrive/PoseToCode - Top Level Folder/RealWorldStudyData/")
```
