```python
# Perform two sample t-test using SciPy
# install bioinfokit as !pip install bioinfokit
!pip install bioinfokit
from scipy import stats as st
from bioinfokit.analys import get_data
import pandas as pd
```

```python
def getScore(response):
  if response == "Strongly Agree":
    return 5

  elif response == "Agree":
    return 4

  elif response == "Neither Agree Nor Disagree":
    return 3

  elif response == "Disagree":
    return 2

  elif response == "Strongly Sisagree":
    return 1

  else:
    return 0
```

```python
# load dataset as pandas dataframe
df_p2c_survey = pd.read_csv('/content/drive/MyDrive/PoseToCode - Top Level Folder/DataAnalysis/export_p2c_cleaned.csv', usecols=["Q3"])
#df_p2c_survey = df_p2c_survey.drop([0, 1])
df_p2c_survey.insert(0, "Activity", ["p2c", "p2c", "p2c", "p2c", "p2c", "p2c", "p2c", "p2c", "p2c", "p2c"], True)
print(df_p2c_survey)
```

```python
df_codeorg_survey = pd.read_csv('/content/drive/MyDrive/PoseToCode - Top Level Folder/DataAnalysis/export_codeorg_cleaned.csv', usecols=["Q3"])
#df_codeorg_survey = df_codeorg_survey.drop([0, 1])
df_codeorg_survey.insert(0, "Activity", ["code.org", "code.org", "code.org", "code.org", "code.org", "code.org", "code.org", "code.org", "code.org", "code.org"], True)
print(df_codeorg_survey)
```

```python
frames = [df_p2c_survey, df_codeorg_survey]
result = pd.concat(frames)
result = result.dropna()
display(result)
```

```python
for index, row in result.iterrows():
    row['Q3'] = getScore(row['Q3'])
display(result)
```

```python
a = result.loc[result['Activity'] == 'p2c', 'Q3'].to_numpy()
b = result.loc[result['Activity'] == 'code.org', 'Q3'].to_numpy()
st.ttest_ind(a=a, b=b, equal_var=True)
```

```python
from bioinfokit.analys import stat
res = stat()
# for unequal variance t-test (Welch's t-test) set evar=False
res.ttest(df=result, xfac="Activity", res="Q3", test_type=2, evar=False)
print(res.summary)
```
