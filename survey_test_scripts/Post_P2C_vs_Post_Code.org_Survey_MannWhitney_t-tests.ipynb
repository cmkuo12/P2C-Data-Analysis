```python
import pandas as pd
import scipy.stats as stats
import math
```

```python
def getScore(response):
  if response == "Strongly Agree":
    return 5

  elif response == "Agree":
    return 4

  elif response == "Neither Agree Nor Disagree":
    return 3

  elif response == "Disagree":
    return 2

  elif response == "Strongly Sisagree":
    return 1

  else:
    return 0
```

```python
# load dataset as pandas dataframe
df_p2c_survey = pd.read_csv('/content/drive/MyDrive/PoseToCode - Top Level Folder/DataAnalysis/export_p2c_cleaned.csv', usecols=["Q1_1"])
df_p2c_survey = df_p2c_survey.rename(columns={'Q1_1': 'P2C'})
print(df_p2c_survey)
```

```python
df_codeorg_survey = pd.read_csv('/content/drive/MyDrive/PoseToCode - Top Level Folder/DataAnalysis/export_codeorg_cleaned.csv', usecols=["Q1_1"])
df_codeorg_survey = df_codeorg_survey.rename(columns={'Q1_1': 'Code.org'})
print(df_codeorg_survey)
```

```python
df_merged = pd.concat([df_p2c_survey, df_codeorg_survey], axis=1)
display(df_merged)
```

```python
for index, row in df_merged.iterrows():
    row['P2C'] = getScore(row['P2C'])
    row['Code.org'] = getScore(row['Code.org'])
display(df_merged)
```

```python
import scipy.stats as stats
# perform two-sided test. You can use 'greater' or 'less' for one-sided test
print("CT_POST QUESTION")
u,p = stats.mannwhitneyu(x=df_merged['P2C'], y=df_merged['Code.org'], alternative = 'two-sided')
print(u)
print(p)
```

```python
n1 = len(df_p2c_survey.index)
n2 = len(df_codeorg_survey.index)
exp_value = (n1 * n2)/2
std_error = math.sqrt((n1*n2*(n1+n2+1))/12)
z = (u - exp_value)/std_error
effect_size = abs(z)/math.sqrt(n1)
print("effect size: " + str(effect_size))
```
