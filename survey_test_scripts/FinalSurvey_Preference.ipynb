```python
import pandas as pd
import scipy.stats as stats
import math
```

```python
# load dataset as pandas dataframe
df_p2c_survey = pd.read_csv('/content/drive/MyDrive/PoseToCode - Top Level Folder/DataAnalysis/export_p2c_cleaned.csv', usecols=["StartDate", "STUID"])
df_p2c_survey = df_p2c_survey.rename(columns={'StartDate': 'StartP2C'})
print(df_p2c_survey)
```

```python
df_codeorg_survey = pd.read_csv('/content/drive/MyDrive/PoseToCode - Top Level Folder/DataAnalysis/export_codeorg_cleaned.csv', usecols=["StartDate", "STUID"])
df_codeorg_survey = df_codeorg_survey.rename(columns={'StartDate': 'StartCodeorg'})
print(df_codeorg_survey)
```

```python
import numpy as np

df_codeorg_first = pd.DataFrame()
df_p2c_first = pd.DataFrame()


df_p2c_survey['StartP2C'] = pd.to_datetime(df_p2c_survey['StartP2C'])
df_codeorg_survey['StartCodeorg'] = pd.to_datetime(df_codeorg_survey['StartCodeorg'])

df_p2c_first = df_p2c_survey.merge(df_codeorg_survey, on='STUID').query('StartP2C < StartCodeorg')
df_codeorg_first = df_p2c_survey.merge(df_codeorg_survey, on='STUID').query('StartP2C > StartCodeorg')
print(df_p2c_first)
print(df_codeorg_first)
```

```python
df_final_survey = pd.read_csv('/content/drive/MyDrive/PoseToCode - Top Level Folder/DataAnalysis/export_final_cleaned.csv', usecols=["Q19", "STUID"])
df_final_survey = df_final_survey.rename(columns={'Q19': 'Preference'})
print(df_final_survey)
```

```python
#code.org first
codeorg_first_participants_list = df_codeorg_first['STUID'].tolist();

df_codeorg_first = df_final_survey[df_final_survey['STUID'].isin(codeorg_first_participants_list)]

print(df_codeorg_first)
```

```python
#p2c first
p2c_first_participants_list = df_p2c_first['STUID'].tolist();

df_p2c_first = df_final_survey[df_final_survey['STUID'].isin(p2c_first_participants_list)]

print(df_p2c_first)
```
