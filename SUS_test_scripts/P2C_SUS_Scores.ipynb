```python
import pandas as pd
df_p2c_survey = pd.read_csv('/content/drive/MyDrive/PoseToCode - Top Level Folder/DataAnalysis/export_p2c_cleaned.csv', usecols=[*range(19, 29), 32])
print(df_p2c_survey)
```

```python
def getScore(response):
  if response == "Strongly agree":
    return 5

  elif response == "Somewhat agree":
    return 4

  elif response == "Neither agree nor disagree":
    return 3

  elif response == "Somewhat disagree":
    return 2

  elif response == "Strongly disagree":
    return 1

  else:
    return 0
```

```python
finalScoreSum = 0;
scores = []
df_sus = pd.DataFrame(columns=['STUID', 'SUS'])

for index in df_p2c_survey.index:
  totalOddScore = 0;
  totalEvenScore = 0;
  
  s = getScore(df_p2c_survey['Q1_1.1'][index])
  totalOddScore += s;

  s = getScore(df_p2c_survey['Q1_2'][index])
  totalEvenScore += s;

  s = getScore(df_p2c_survey['Q1_3'][index])
  totalOddScore += s;

  s = getScore(df_p2c_survey['Q1_4'][index])
  totalEvenScore += s;

  s = getScore(df_p2c_survey['Q1_5'][index])
  totalOddScore += s;

  s = getScore(df_p2c_survey['Q1_6'][index])
  totalEvenScore += s;

  s = getScore(df_p2c_survey['Q1_7'][index])
  totalOddScore += s;

  s = getScore(df_p2c_survey['Q1_8'][index])
  totalEvenScore += s;

  s = getScore(df_p2c_survey['Q1_9'][index])
  totalOddScore += s;

  s = getScore(df_p2c_survey['Q1_10'][index])
  totalEvenScore += s;

  totalOddScore -= 5
  totalEvenScore = 25 - totalEvenScore

  finalScore = (totalEvenScore + totalOddScore)*2.5

  print(finalScore)
  df1 = pd.DataFrame([[df_p2c_survey['STUID'][index], finalScore]], columns=['STUID','SUS'])
  df_sus = pd.concat([df1, df_sus], ignore_index=True)
  scores.append(finalScore)
  finalScoreSum += finalScore;

print("FINAL SUS SCORE: " + str(finalScoreSum / len(df_p2c_survey.index)))
print(df_sus)
df_sus.to_csv('/content/drive/MyDrive/PoseToCode - Top Level Folder/DataAnalysis/export_p2c_sus_scores.csv', index=False)
```

```python
import numpy
print("std: " + str(numpy.std(scores)))
print("var: " + str(numpy.var(scores)))
print("mean: " + str(numpy.mean(scores)))
print("med: " + str(numpy.median(scores)))
```
