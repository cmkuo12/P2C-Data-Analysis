```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/PoseToCode - Top Level Folder/DataAnalysis/export_codeorg_cleaned.csv', usecols=['Q13_1', 'Q13_2', 'Q13_3', 'Q13_4', 'Q13_5', 'Q13_6', 'Q13_7', 'Q13_8', 'Q13_9', 'Q13_10'])

```

```python
def getScore(response):
  if response == "Strongly agree":
    return 5

  elif response == "Somewhat agree":
    return 4

  elif response == "Neither agree nor disagree":
    return 3

  elif response == "Somewhat disagree":
    return 2

  elif response == "Strongly disagree":
    return 1

  else:
    return 0
```

```python
finalScoreSum = 0;
scores = []

for index in df.index:
  totalOddScore = 0;
  totalEvenScore = 0;
  
  s = getScore(df['Q13_1'][index])
  totalOddScore += s;

  s = getScore(df['Q13_2'][index])
  totalEvenScore += s;

  s = getScore(df['Q13_3'][index])
  totalOddScore += s;

  s = getScore(df['Q13_4'][index])
  totalEvenScore += s;

  s = getScore(df['Q13_5'][index])
  totalOddScore += s;

  s = getScore(df['Q13_6'][index])
  totalEvenScore += s;

  s = getScore(df['Q13_7'][index])
  totalOddScore += s;

  s = getScore(df['Q13_8'][index])
  totalEvenScore += s;

  s = getScore(df['Q13_9'][index])
  totalOddScore += s;

  s = getScore(df['Q13_10'][index])
  totalEvenScore += s;

  totalOddScore -= 5
  totalEvenScore = 25 - totalEvenScore

  finalScore = (totalEvenScore + totalOddScore)*2.5

  print(finalScore)
  scores.append(finalScore)
  finalScoreSum += finalScore;

print("FINAL SUS SCORE: " + str(finalScoreSum / len(df.index)))
```

```python
import numpy
print("std: " + str(numpy.std(scores)))
print("var: " + str(numpy.var(scores)))
print("mean: " + str(numpy.mean(scores)))
print("med: " + str(numpy.median(scores)))
```
